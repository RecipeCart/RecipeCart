```python
# STEP 1 - Prepare a helper function to iterate through data in batches
def get_batch_with_cursor(collection_name, batch_size, cursor=None):
    # First prepare the query to run through data
    query = (
        client.query.get(
            collection_name,         # update with your collection name
            ["name"] # update with the required properties
        )
        .with_additional(["id"])
        .with_limit(batch_size)
    )

    # Fetch the next set of results
    if cursor is not None:
        result = query.with_after(cursor).do()
    # Fetch the first set of results
    else:
        result = query.do()
    return result["data"]["Get"][collection_name]

# STEP 2 - Iterate through the data
cursor = None
ingredients_map = {}
while True:
    # Get the next batch of objects
    next_batch = get_batch_with_cursor("IngredientType", 100, cursor)

    # Break the loop if empty – we are done
    if len(next_batch) == 0:
        break

    # Here is your next batch of objects
    # print(next_batch)
    for entry in next_batch:
      ingredients_map[entry['name']] = entry['_additional']['id']


    # Move the cursor to the last returned uuid
    cursor=next_batch[-1]["_additional"]["id"]
```

```python
data_objs_recipes = []
for row in df.itertuples():
  slice_list = [ingredient_slicer.IngredientSlicer(i.lower()).to_json() for i in eval(row[6])]
  food_list = []
  for i in slice_list:
    if i['food'] == '':
      continue
    try:
      food_list += [ingredients_map[i['food']]]
    except KeyError:
      ingredients_lst = model.extract_foods(i['standardized_ingredient'])
      for ingredients in ingredients_lst:
        for ingredient in ingredients['Product']+ingredients['Ingredient']:
          try:
            food_list += [ingredients_map[ingredient['text']]]
          except KeyError:
            #add i['food'] to IngredientTypes
            uuid = client.data_object.create(
                class_name="IngredientType",
                data_object = {
                    "name": i['food'],
                    "primary_category": 'unknown',
                    "secondary_category": 'unknown',
                }
            )
            ingredients_map[ingredient['text']] = uuid
            food_list += [ingredients_map[ingredient['text']]]

      # ingredients_map[i['food']] = uuid
      # food_list += [ingredients_map[i['food']]]
  #2. Title, 4. Instructions, 6. Ingredients, using map on 6 for uuid
  data_objs_recipes += [{'title':row[2], 'instructions': row[4], 'ingredients_sliced': eval(row[6]), 'hasIngredientType': food_list, 'rating': [0,0,0,0,0], 'img_name':row[5]}]
```
